## IMPORTING LIBRARIES

```python
#libraries
import os
from alpha_vantage.foreignexchange import ForeignExchange
from datetime import datetime

import matplotlib.pyplot as plt
from dotenv import load_dotenv
dotenv_path = os.path.join(os.getcwd(), '.env')
load_dotenv(dotenv_path)
alpha = os.environ['ALPHA']


from pprint import pprint
import pandas as pd
import numpy as np
```

```python
#API Key
fx = ForeignExchange(key='alpha')
```

## GETTING DAILY CURRENCY EXCHANGE DATA

# EURO - EUR

```python
#EUR-USD DATA
data, meta_data = fx.get_currency_exchange_daily(from_symbol='EUR',to_symbol='USD', outputsize='full')

EUR_df = pd.DataFrame()

for date in data:
    temp = pd.DataFrame(
    {
        "Date" : date,
        "EUR" : [data[date]['4. close']]
    }
    )
    
    EUR_df = pd.concat([EUR_df, temp], ignore_index=True)
    
EUR_df['Date'] = pd.to_datetime(EUR_df['Date'])
EUR_df['EUR'] = EUR_df['EUR'].astype('float')
EUR_df.set_index('Date', inplace = True)
EUR_df.sort_index(ascending=False)
EUR_df.head()
```

```python
EUR_df.plot()
plt.tight_layout()
plt.title('Daily close value for EUR - USD')
plt.grid()
plt.show()
```

## COMPARISON FOREX DATA WITH BTC

```python
#Read in Bitcoin Data 

btc_df = pd.read_csv("Data\BTC-USD.csv", index_col="Date", parse_dates=True, infer_datetime_format=True)
btc_df = btc_df.drop(columns=["Open", "High", "Low", "Adj Close", "Volume"]) #Keep relevant columns 
btc_df = btc_df.rename(columns={'Close':'BTC'}) #rename close column to BTC

#Read in Ethereum Data 
eth_df = pd.read_csv("Data\ETH-USD.csv", index_col="Date", parse_dates=True, infer_datetime_format=True)
eth_df = eth_df.drop(columns=["Open", "High", "Low", "Adj Close", "Volume"]) #Keep relevant columns 
eth_df = eth_df.rename(columns={'Close':'ETH'}) #rename close column to ETH

#Read in Gold Futures Prices 
gold_df = pd.read_csv("Data\Gold Futures Prices.csv", index_col="Date", parse_dates=True, infer_datetime_format=True)
gold_df = gold_df.drop(columns=["Vol.", "Open", "High", "Low", "Change %"]) #Keep relevant columns 
gold_df = gold_df.rename(columns={'Price':'GOLD'}) #rename close column to GOLD
```

```python
#Concatenate Data

combined_df = pd.concat([btc_df, eth_df, gold_df] ,axis="columns", join="inner")
combined_df.head()
```

```python
#Change Data type of "Gold" to float instead of object 

combined_df['GOLD'] = combined_df['GOLD'].str.replace(',', '')
combined_df['GOLD'] = combined_df['GOLD'].astype('float')

combined_returns = combined_df.pct_change().dropna()
```
