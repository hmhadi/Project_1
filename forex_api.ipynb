## IMPORTING LIBRARIES

```python
#libraries
import os
from alpha_vantage.foreignexchange import ForeignExchange
from datetime import datetime

import matplotlib.pyplot as plt
from dotenv import load_dotenv
dotenv_path = os.path.join(os.getcwd(), '.env')
load_dotenv(dotenv_path)
alpha = os.environ['ALPHA']


from pprint import pprint
import pandas as pd
```

```python
#API Key
fx = ForeignExchange(key='alpha')
```

## GETTING DAILY CURRENCY EXCHANGE DATA
BASE CURRENCY : USD







# EURO - EUR

```python
#USD_EUR DATA
data, meta_data = fx.get_currency_exchange_daily(from_symbol='USD',to_symbol='EUR', outputsize='full')

EUR_df = pd.DataFrame()

for date in data:
    temp = pd.DataFrame(
    {
        "Date" : date,
        "EUR" : [data[date]['4. close']]
    }
    )
    
    EUR_df = pd.concat([EUR_df, temp], ignore_index=True)
EUR_df['Date'] = pd.to_datetime(EUR_df['Date'])
EUR_df['EUR'] = EUR_df['EUR'].astype('float')
EUR_df.set_index('Date', inplace = True)
EUR_df.sort_index(ascending=False)
EUR_df.head()
```

```python
EUR_df.plot()
plt.tight_layout()
plt.title('Daily close value for USD - EUR')
plt.grid()
plt.show()
```
